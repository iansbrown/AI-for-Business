```python
import pandas as pd
import numpy as np
import os
import glob
from PIL import Image
import shutil

df = pd.read_csv('Data Workshop 2 May 1 2023.csv')
photos=glob.glob('/users/ianbrown/AIBusiness/recipe1/*.jpg')
print (len(photos))
try:
    os.makedirs('recipe1tagged')
except FileExistsError:
    pass

dest_dir = os.getcwd()+'/recipe1tagged'
for file in photos:
    parsed = file.split('_')
    serial=parsed[2]
    filename=file.split('/')[-1]
    print (filename)
    row_num = df[df['MESScheduleUniqueID'] == serial].index.to_numpy()
    print (row_num[0])
    #print(df.loc[row_num,['InspectStationWeighCheckData','VialEmptyWeight','InspectStationWeighCheckPassFail','InspectStationWeighCheckLowerLimit','InspectStationWeighCheckUpperLimit']]) 
    pf=df._get_value(row_num[0],'InspectStationWeighCheckPassFail')
    print (pf)
    if pf==1:
        shutil.copy(file,dest_dir)
        os.rename(dest_dir+'/'+filename, dest_dir+'/proper_fill'+serial+'.jpg')
    elif pf==2:
        weight=df._get_value(row_num[0],'InspectStationWeighCheckData')
        upper=df._get_value(row_num[0],'InspectStationWeighCheckUpperLimit')
        lower=df._get_value(row_num[0],'InspectStationWeighCheckLowerLimit')
        if weight > upper:
            shutil.copy(file,dest_dir)
            os.rename(dest_dir+'/'+filename, dest_dir+'/over_fill'+serial+'.jpg')
        elif weight < lower:
            shutil.copy(file,dest_dir)
            os.rename(dest_dir+'/'+filename, dest_dir+'/under_fill'+serial+'.jpg')
    else:
        shutil.copy(file,dest_dir)
        os.rename(dest_dir+'/'+filename, dest_dir+'/uncategorized'+serial+'.jpg')       
```
