```python
from azureml.opendatasets import NycTlcGreen
import pandas as pd
from datetime import datetime
from dateutil.relativedelta import relativedelta
```

```python
green_taxi_df = pd.DataFrame([])
start = datetime.strptime("1/1/2016","%m/%d/%Y")
end = datetime.strptime("1/31/2016","%m/%d/%Y")
for sample_month in range(12):
    temp_df_green = NycTlcGreen(start + relativedelta(months=sample_month), end + relativedelta(months=sample_month)) \
        .to_pandas_dataframe()
    green_taxi_df = green_taxi_df.append(temp_df_green.sample(2000))
green_taxi_df.head(10)
```

```python
columns_to_remove = ["lpepDropoffDatetime", "puLocationId", "doLocationId", "extra", "mtaTax","improvementSurcharge", "tollsAmount", "ehailFee", 
"tripType", "rateCodeID", "storeAndFwdFlag", "paymentType", 
"fareAmount", "tipAmount"   ]
for col in columns_to_remove:
    green_taxi_df.pop(col)
green_taxi_df.head(10)
```

```python
df = green_taxi_df
df['month'] = pd.DatetimeIndex(df['lpepPickupDatetime']).month
df['dayofweek'] = pd.DatetimeIndex(df['lpepPickupDatetime']).dayofweek
df['hourofday'] = pd.DatetimeIndex(df['lpepPickupDatetime']).hour
```

```python
df.describe()
```

```python
final_df = df.query("tripDistance>=0.25 and tripDistance<31")
final_df = df.query("passengerCount>0 and totalAmount>0")
columns_to_remove_for_training = ["pickupLongitude", "pickupLatitude", "dropoffLongitude", "dropoffLatitude","lpepPickupDatetime"]
for col in columns_to_remove_for_training:
    final_df.pop(col)
```

```python
final_df.describe()
```

```python
firstten = final_df.head(10)
print("First 10 rows of the DataFrame:")
print(firstten)
```

```python
import matplotlib.pyplot as plt
import matplotlib
# Create histogram on column
%matplotlib inline
 
plt.hist(final_df['passengerCount'], color='blue', edgecolor='black', bins=int(6))
plt.xlabel('passengers per trip')
plt.ylabel('number of trips')
plt.title('passenger counts')
plt.show()

plt.hist(final_df['tripDistance'], color='red', edgecolor='black', bins=int(100))
plt.xlim((0,25))
plt.xlabel('passengers per trip')
plt.ylabel('number of trips')
plt.title('trip distance 1')
plt.show()

plt.hist(final_df['tripDistance'], color='red', edgecolor='black', bins=int(100))
plt.xlim((25,50))
plt.ylim((0,5))
plt.xlabel('passengers per trip')
plt.ylabel('number of trips')
plt.title('trip distance 2')
plt.show()

plt.hist(final_df['totalAmount'], color='green', edgecolor='black', bins=int(100))
plt.xlabel('amount per trip')
plt.ylabel('number of trips')
plt.title('total amount')
plt.show()

plt.hist(final_df['month'], color='orange', edgecolor='black', bins=int(12))
plt.ylim((1950,2000))
plt.xlabel('month of trip')
plt.ylabel('number of trips')
plt.title('trips by month')
plt.show()

plt.hist(final_df['dayofweek'], color='yellow', edgecolor='black', bins=int(7))
plt.xlabel('day of week')
plt.ylabel('number of trips')
plt.title('trips by day of the week')
plt.show()

plt.hist(final_df['hourofday'], color='cyan', edgecolor='black', bins=int(24))
plt.xlabel('hour of day')
plt.ylabel('number of trips')
plt.title('trips by hour of day')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

# Separate features and labels
X, y = final_df[['vendorID','passengerCount','tripDistance','month','dayofweek','hourofday']].values, final_df['totalAmount'].values
print('Features:',X[:10], '\nLabels:', y[:10], sep='\n')

# Split data 75%-25% into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

print ('Training Set: %d rows\nTest Set: %d rows' % (X_train.shape[0], X_test.shape[0]))
```

```python
# Train the LR model
from sklearn.linear_model import LinearRegression

# Fit a linear regression model on the training set
modelLR = LinearRegression().fit(X_train, y_train)
print (modelLR)

```

```python
import numpy as np

predictionsLR = modelLR.predict(X_test)
np.set_printoptions(suppress=True)
#print('Linear Regression:')
#print('Predicted labels: ', np.round(predictionsLR)[:10])
#print('Actual labels   : ' ,y_test[:10])

```

```python
from sklearn.metrics import mean_squared_error, r2_score
print ('Linear Regression:')
mse = mean_squared_error(y_test, predictionsLR)
print("MSE:", mse)
rmse = np.sqrt(mse)
print("RMSE:", rmse)
r2 = r2_score(y_test, predictionsLR)
print("R2:", r2)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import export_text

# Train the Decision tree model
modelDT = DecisionTreeRegressor().fit(X_train, y_train)
print (modelDT, "\n")

predictionsDT = modelDT.predict(X_test)
np.set_printoptions(suppress=True)
#print('Descision Tree:')
#print('Predicted labels: ', np.round(predictionsDT)[:10])
#print('Actual labels   : ' ,y_test[:10])
print ('Decision Tree:')
mse = mean_squared_error(y_test, predictionsDT)
print("MSE:", mse)
rmse = np.sqrt(mse)
print("RMSE:", rmse)
r2 = r2_score(y_test, predictionsDT)
print("R2:", r2)
```

```python
from sklearn.ensemble import RandomForestRegressor

# Train the model
modelRF = RandomForestRegressor().fit(X_train, y_train)
print (modelRF, "\n")

# Evaluate the model using the test data
predictionsRF = modelRF.predict(X_test)
print('Random Forrest:')
#print('Predicted labels: ', np.round(predictionsRF)[:10])
#print('Actual labels   : ' ,y_test[:10])
mse = mean_squared_error(y_test, predictionsRF)
print("MSE:", mse)
rmse = np.sqrt(mse)
print("RMSE:", rmse)
r2 = r2_score(y_test, predictionsRF)
print("R2:", r2)
```
